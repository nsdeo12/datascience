```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

### HISTOGRAMS

```python
#See how total bill is distributed with kde (kernel density estimation)
# sns.distplot(tips['total_bill'])

#See how total bill is distributed without kde (kernel density estimation)
# sns.distplot(tips['total_bill'],kde=False)

#See how total bill is distributed without kde (kernel density estimation) and with some bin size
#try with 30 too.with 100 almost all price points are visible
sns.distplot(tips['total_bill'],kde=True,bins=100)
```

### BIVARIANT PLOTS (COMBINE TWO DISTRIBUTION PLOTS)

```python
# total bill along x and tip along y
# The scatter plot is linearly increasing
# Which implies as the bill increases so does the tip
sns.jointplot(x='total_bill',y='tip',data=tips)

# A hexagonal representation( graph 2) can be done with kind = 'hex'
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')

# A linear regression( graph 3) can be done with kind = 'reg'
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')

# A kde plot( graph 4) can be done with kind = 'kde',shows where density is the most 
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

### Pair plot (a pair wise plot between every element in the dataset)

```python
#Its a quick way to visualize your data
sns.pairplot(tips,hue='sex')
```

### Rug plot

```python
sns.rugplot(tips['total_bill'])
```
