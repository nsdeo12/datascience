### Plot Appearance

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='indigo',linewidth=3,alpha=0.8,linestyle='--',marker='o',
        markersize=20,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='red') 
#alpha is transparency and  linewidth (can be lw) 
#linestyle(ls) can be -- marker shows where the datas are
#refer for marker size too in documentatio ,try o as marker type
```

#### Xlimit(show plot between two values),Get a portion of the plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlim([0,2])
ax.plot(x,y,color='indigo',linewidth=3,alpha=0.8,linestyle='--',marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='red')
```
