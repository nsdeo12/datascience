```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
y
```

## Functional method matplotlib


```python
plt.plot(x,y)
# OR

plt.xlabel("--------X----->")
plt.ylabel("--------Y----->")
plt.title('Graph')
plt.plot(x,y)
# plt.show()
```

#### SUBPLOT

```python
#subplot(nrows, ncols, index, **kwargs) index is the plot number
plt.subplot(1,2,1)
# r stands for red 
plt.plot(x,y,'r')

#for second plot in blue color
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object oriented method

```python
fig = plt.figure()
# left bottom width and height are the four parameters
mainPlot=fig.add_axes([0.1,0.1,1.1,1.1])
mainPlot.set_xlabel('-------Main X----->>>>>>')
mainPlot.set_ylabel('-------Main Y----->>>>>>')

axes1=fig.add_axes([0.20,0.3,0.3,0.5])
axes2=fig.add_axes([0.7,0.3,0.3,0.5])

axes1.set_xlabel('----X1---->>>>')
axes1.set_ylabel('----Y1---->>>>')


axes2.set_xlabel('----X2---->>>>')
axes2.set_ylabel('----Y2---->>>>')
axes1.plot(x,y)
axes2.plot(x,y)
```
