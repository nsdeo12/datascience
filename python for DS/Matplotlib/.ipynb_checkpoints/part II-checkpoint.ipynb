```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
y
```

```python
fig=plt.figure()
plot1=fig.add_axes([0.1,0.1,0.8,0.8])
plot1.set_xlabel("---------------X---->")
plot1.set_ylabel("---------------Y---->")
plot1.plot(x,y)
```

## Define multiple plots


```python
# step 1 : define two plots
fig,axes=plt.subplots(nrows=1,ncols=2)

```

```python
#step 2 :add graph
fig,axes=plt.subplots(nrows=1,ncols=2)
for item in axes:
    item.plot(x,y)
```

### iterate through the plots

```python
#get first graph
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#Guess what????
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].plot(y,x)
```

### Figure size and DPI

```python
# define with figsize and dpi
fig,axes=plt.subplots(nrows=2,ncols=1,dpi=300,figsize=(8,2))
#compact the figure
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(6,6))
axes[0].plot(x,y,'r')
axes[0].set_xlabel("---ding--->")
axes[0].set_ylabel("---dong--->")

axes[1].plot(y,x**2,'g')
plt.tight_layout()

#save the figure with proper extension
fig.savefig('myFigure.svg',dpi=200)
```

### Legend
#### requires label keyword

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x*2,label='x vs 2x')
ax.plot(x,x**2,label='x vs xSquared')
#change location code (https://matplotlib.org/api/legend_api.html)
ax.legend(loc=1) # at top left corner
ax.legend(loc=(0.1,0.2))
```
