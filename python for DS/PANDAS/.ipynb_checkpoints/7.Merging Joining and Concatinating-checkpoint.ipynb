```python
import numpy as np
import pandas as pd
```

```python
data1={'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']}

#notice index is passed too ,else all will be 0,1,2,3 only

df1=pd.DataFrame(data1,index=[0,1,2,3])
data2={'A':['A4','A5','A6','A7'],'B':['B4','B5','B6','B7'],'C':['C4','C5','C6','C7'],'D':['D4','D5','D6','D7']}
df2=pd.DataFrame(data2,index=[4,5,6,7])
data3={'A':['A8','A9','A10','A11'],'B':['B8','B9','B10','B11'],'C':['C8','C9','C10','C11'],'D':['D8','D9','D10','D11']}
df3=pd.DataFrame(data3,index=[8,9,10,11])
print(df1,'\n','\n',df2,'\n','\n',df3)
```

### CONCAT 
#### (should be passed in [])


```python

#concatinate (should be passed in [])
pd.concat([df1,df2,df3])
```

```python


#concatinate (should be passed in [])
pd.concat([df1,df2,df3],axis=1)
```

### MERGING

```python
left=pd.DataFrame({'key':['K0','K1','K2','K3'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key':['K0','K1','K2','K3'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']})
left
```

```python
right
```

```python
### Merging
pd.merge(left,right)

#in Details
pd.merge(left,right,how='inner',on='key')
```

```python
left1=pd.DataFrame({'key1':['K0','K0','K1','K2'],'key2':['K0','K1','K0','K1'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
right1=pd.DataFrame({'key1':['K0','K1','K1','K2'],'key2':['K0','K0','K0','K0'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']})
left1
```

```python
right1
```

```python
pd.merge(left1,right1,how='outer',on=['key1','key2'])
```
